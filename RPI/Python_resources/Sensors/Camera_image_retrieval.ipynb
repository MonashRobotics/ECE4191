```python
import cv2
from matplotlib import pyplot as plt
```

### Getting frames from the Raspberry PI Camera

On ubuntu, I needed to include `start_x=1` inside the file `/boot/firmware/config.txt`, then reboot the pi to get the camera attached. If you're using Raspbian, check out [picamera](https://picamera.readthedocs.io/en/release-1.0/quickstart.html).

```python
cap = cv2.VideoCapture(0)
cap.set(cv2.CAP_PROP_FRAME_WIDTH,640)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT,480)
cap.set(cv2.CAP_PROP_FPS, 40)
```

```python
ret = False
while (not ret):
    ret, frame = cap.read()

plt.imshow(frame[:,:,[2,1,0]])
plt.show()
```

### But why is the picture purple?
The NoIR Camera has no infrared filter on the lens. This helps with low-light environment photography (try taking some pictures in the dark).
